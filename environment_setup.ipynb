```python
# Import libraries
import gym
import numpy as np
import matplotlib.pyplot as plt
import gym_sokoban

#from gym_sokoban.envs.sokoban_env import SokobanEnv
```

## Create and Test Environment


```python
# Create Sokoban environment
env = gym.make('Sokoban-v0')

# Reset environment
obs = env.reset()

print("environment created")
print(f"Observation shape: {obs.shape}")
print(f"Action space: {env.action_space}")
print(f"Number of actions: {env.action_space.n}")
```

```python
# Inspect the environment state
print("Room state shape:", env.room_state.shape)
print("Room fixed shape:", env.room_fixed.shape)
print("\nRoom state values:")
print("0 = empty, 1 = wall, 2 = goal, 3 = box on goal, 4 = box, 5 = player")
print("\nRoom fixed values:")
print("0 = wall, 1 = empty, 2 = goal")
```

```python
# Visualize the initial state
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Room state (dynamic elements: boxes, player)
axes[0].imshow(env.room_state, cmap='viridis')
axes[0].set_title('Room State (Dynamic)')
axes[0].set_xlabel('Column')
axes[0].set_ylabel('Row')

# Room fixed (static elements: walls, goals)
axes[1].imshow(env.room_fixed, cmap='viridis')
axes[1].set_title('Room Fixed (Static)')
axes[1].set_xlabel('Column')
axes[1].set_ylabel('Row')

plt.tight_layout()
plt.show()
```

```python

obs = env.reset()

done = False

while not done:
    env.render(mode="human")   # shows a popup window with the Sokoban grid

    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
```

## Test Environment Actions


```python
# Test a few random actions
print("Testing environment actions...\n")

for step in range(5):
    # Sample a random action
    action = env.action_space.sample()
    
    # Take step
    obs, reward, done, info = env.step(action)
    
    print(f"Step {step + 1}:")
    print(f"  Action: {action} (0=up, 1=down, 2=left, 3=right)")
    print(f"  Reward: {reward}")
    print(f"  Done: {done}")
    print(f"  Info: {info}")
    print()
    
    if done:
        print("Episode finished!")
        break
```

## Environment Information


```python
# Get environment information
print("Environment Information:")
print("=" * 50)
print(f"Action space: {env.action_space}")
print(f"Number of actions: {env.action_space.n}")
print(f"Observation space: {env.observation_space}")
print(f"Observation shape: {obs.shape}")
print(f"Room dimensions: {env.room_state.shape}")
print("\nAction mapping:")
print("  0 = Move Up")
print("  1 = Move Down")
print("  2 = Move Left")
print("  3 = Move Right")
print("\nReward structure:")
print("  +10 for solving the puzzle")
print("  -0.1 per move")
print("  +1 for box on goal")
print("  -1 for box off goal")
```

## Reset and Close Environment


```python
# Reset environment
obs = env.reset()
print("Environment reset successfully!")

# Close environment when done
env.close()
print("Environment closed.")
```
